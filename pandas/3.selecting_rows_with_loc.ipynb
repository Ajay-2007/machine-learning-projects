### Index Operator loc (label based indexing)

#### Selecting Rows with loc

```python
import pandas as pd
```

```python
summer = pd.read_csv('./datasets/summer.csv')
```

```python
summer.head()
```

```python
summer.iloc[2]
```

```python
summer = pd.read_csv('./datasets/summer.csv', index_col='Athlete')
```

```python
summer.head()
```

```python
summer.iloc[2]
```

```python
summer.loc['DRIVAS, Dimitrios']
```

```python
summer.loc['PHELPS, Michael']
```

```python
summer.loc[
    PHELPS, Michael'].iloc[0]
```

### Slicing Rows and Columns with loc

```python
summer.head()
```

```python
summer.loc['PHELPS, Michael']
```

```python
summer.loc['PHELPS, Michael', 'Medal']
```

```python
summer.loc['PHELPS, Michael', ['Medal', 'Event']]
```

```python
summer.loc[['PHELPS, Michael', 'LEWIS, Carl'], ['Medal', 'Event']]
```

```python
summer.head()
```

```python
summer.loc[:'CHASAPIS, Spiridon']
```

```python
# KeyError: "Cannot get right slice bound for non-unique label: 'PHELPS, Michael'"
# summer.loc[:'PHELPS, Michael']
```

```python
# KeyError: "Cannot get left slice bound for non-unique label: 'HAJOS, Alfred'"
# summer.loc['HAJOS, Alfred':]
```

#### label based slicing only works with unique label

```python
summer.head(20)
```

```python
summer.loc['DRIVAS, Dimitrios' : 'BLAKE, Arthur']
```

```python
summer.loc['DRIVAS, Dimitrios', 'Year' : 'Sport']
```

```python
summer.loc[['PHELPS, Michael' ,'DUCK, Donald']]
```

```python
summer.loc['PHELPS, Michael', ['Year', 'Age']] 
```
